##Task 6:Find the outliers and replace the outliers

```python
sns.boxplot(dataset['CreditScore'])
```

```python
#quantile
qnt=dataset.quantile(q=(0.75,0.25))
qnt
```

```python
#IQR calculations
iqr=qnt.loc[0.75]-qnt.loc[0.25]
iqr
```

```python
#lower extreme value
lower=qnt.loc[0.25]-1.5*iqr
lower
```

```python
#upper extreme value
upper=qnt.loc[0.75]+1.5*iqr
upper
```

```python
#mean value
dataset.mean()
```

```python
#replacing the outliers
dataset['CreditScore']=np.where(dataset['CreditScore']<3.830000e+02,6.505288e+02,dataset['CreditScore'])
```

```python
sns.boxplot(dataset['CreditScore'])
```
