##Task 6:Find the outliers and replace them outliers

```python
sns.boxplot(dataset['Diameter'])
```

```python
#quantile
qnt=dataset.quantile(q=(0.75,0.25))
qnt
```

```python
#IQR calculations
iqr=qnt.loc[0.75]-qnt.loc[0.25]
iqr
```

```python
#lower extreme value
lower=qnt.loc[0.25]-1.5*iqr
lower
```

```python
#upper extreme value
upper=qnt.loc[0.75]+1.5*iqr
upper
```

```python
#mean value
dataset.mean()
```

```python
#replacing the outliers
dataset['Diameter']=np.where(dataset['Diameter']< 0.15500,0.407881,dataset['Diameter'])
```

##Task 8: Split the data into dependent and independent variables

##Task 10: Split the data into training and testing

```python
#train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=True)
```

```python
y_train.shape
```

#Task 11: Build the Model

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
linear_model=LinearRegression()
```

```python
lasso_model=Lasso(alpha=0.01,normalize=True)
```

```python
ridge_model=Ridge(alpha=0.01,normalize=True)
```

##Task 12: Train the Model

```python
#training the model
linear_model.fit(x_train,y_train)
```

```python
lasso_model.fit(x_train,y_train)
```

```python
ridge_model.fit(x_train,y_train)
```

##Task 13: Test the Model

```python
#prediction
linear_pred=linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
#accuracy score for linear regression
acc=r2_score(y_test,linear_pred)
acc
```
